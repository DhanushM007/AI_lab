```python
import numpy as np

data = np.array([
    [1200, 200],
    [1500, 250],
    [1700, 200],
    [2100, 400],
    [2300, 450],
    [2500, 500]
])

x_train = data[:4, 0].reshape(-1, 1)
y_train = data[:4, 1]

x_test = data[4:, 0].reshape(-1, 1)
y_test = data[4:, 1]

def knn_regression(x_train, y_train, x_test, k=3):
    predictions = []
    
    for test_point in x_test:
        distances = np.sqrt(np.sum((x_train - test_point) ** 2, axis=1))
        nearest_indices = np.argsort(distances)[:k]
        nearest_values = y_train[nearest_indices]
        prediction = np.mean(nearest_values)
        predictions.append(prediction)
        
    return np.array(predictions)

y_pred = knn_regression(x_train, y_train, x_test, k=3)

for size, actual, predicted in zip(x_test.flatten(), y_test, y_pred):
    print(f"House Size: {size}, Actual Price: {actual}, Predicted Price: {predicted:.2f}")

mse = np.mean((y_test - y_pred) ** 2)
print("\nMean Squared Error (MSE):", mse)
```
