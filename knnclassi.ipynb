```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from collections import Counter

data = load_iris()
x_train, x_test, y_train, y_test = train_test_split(
    data.data, data.target, test_size=0.3, random_state=42
)

def knn_classify(x_train, y_train, x_test, k=3):
    predictions = []
    for test_point in x_test:
        distances = [np.sqrt(np.sum((test_point - train_point)**2)) for train_point in x_train]
        nearest = [y_train[i] for i in np.argsort(distances)[:k]]
        predictions.append(Counter(nearest).most_common(1)[0][0])
    return predictions

preds = knn_classify(x_train, y_train, x_test, k=3)

correct = 0
for i in range(len(preds)):
    if preds[i] == y_test[i]:
        print(f"Test sample {i+1}: Correct (Predicted:{preds[i]}, Actual:{y_test[i]})")
        correct += 1
    else:
        print(f"Test sample {i+1}: Wrong (Predicted:{preds[i]}, Actual:{y_test[i]})")

print(f"\nAccuracy: {correct/len(y_test)*100:.2f}%")
```
